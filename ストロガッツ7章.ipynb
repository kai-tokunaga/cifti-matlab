```python

def twodimentional_polar_phase_space(p,r,fp,fr,g,h):
    P,R = np.meshgrid(p,r)
    u = fp(P,R)   
    v = fr(P,R)
    ax.streamplot(P,R,u,v)
    ax.set_ylim(0,np.max(r))
    try:
        ax.plot(p,g(p,0),color='g')
    except TypeError:
        ax.plot(g(0,r),r,color='g')
    except ValueError:
        ax.plot(g(0,r),r,color='g')
    try:
        ax.plot(p,h(p,0),color='g')
    except ValueError:
        ax.plot(h(0,r),r,color='g')
    except TypeError:
        ax.plot(h(0,r),r,color='g')
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'},figsize=(30,7))
r= np.linspace(0,2,100)
p = np.linspace(0,2*np.pi,100)
def fr(p,r):
    return r*(1-r**2)
def fp(p,r):
    return (p+1)/(p+1)
def g(p,r):
    return np.zeros(len(p))
def h(p,r):
    return np.ones(len(p))
twodimentional_polar_phase_space(p,r,fp,fr,g,h)
```
